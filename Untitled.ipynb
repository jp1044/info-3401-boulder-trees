```python
def dic(c):
    d = {}
    for i in c:
        try:
            d[i]+=1
        except:
            d[i]=1
    import operator
    d = sorted(d.items(), key=operator.itemgetter(1),reverse = True)
    return d
def dic2(c):
    a = {}
    for i in c:
        try:
            a[i]+=1
        except:
            a[i]=1
    import operator
    a = sorted(a.items(), key=operator.itemgetter(1),reverse = True)
    t=0
    for i in a:
        t+=i[1]
    d = []
    for i in a:
        d.append([i[0],float(i[1])/t])
    return d
```

```python
dic2(df['COMMONNAME'])
```
